```python
games = pd.read_csv('../data/review.csv')
print(games.shape)
games.head()
```

```python
games.isna().sum()
```

```python
data = games[['app_id', 'app_name', 'review_text']]
print(data.shape)
data.head()
```

```python
review_df = pd.DataFrame(data)
reviews_df = review_df.drop_duplicates(subset=['app_id'], keep='last')
print(reviews_df.shape)
reviews_df.head()
```

```python
reviews_df.isna().sum()
```

```python
print(reviews_df)
```

```python
reviews_df.rename(columns={"app_id": "appid", "review_text": "review"}, inplace=True)
print(reviews_df.shape)
reviews_df.head()
```

```python
reviews_df.dropna(inplace=True)
```

```python
print(reviews_df.shape)
reviews_df.head()
```

```python
reviews_df.to_csv('../data/reviews.csv', index = False)
```
